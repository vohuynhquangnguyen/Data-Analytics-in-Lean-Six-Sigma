## Questions:
### 1. What are the average and standard deviation of the total time?
**Solution:** 

Our workflow is as follows: _Total Time_ &#8594; Numerical (as in Data Type) &#8594; Descriptive Statistics

In Minitab, go to `Stat` >> `Basic Statistics` >> `Display Descriptive Stastistics`. In the popup dialog, select _Total Time_ as our `Variables` option.


1. The average (mean) of _Total Time_ is 9.422. Thus, it takes approximately 9.42 (days) in average to handle a reclaim.
2. The standard deviation of _Total Time_ is 12.661 (days).

### 2. Use the quartiles to give an estimate of how many reclaims are handled within 15 days.
**Solution:** 

Our workflow is as follows: _Total Time_ &#8594; Numerical (as in Data Type) &#8594; Descriptive Statistics.

In Minitab, go to `Stat` >> `Basic Statistics` >> `Display Descriptive Stastistics`. In the popup dialog, select _Total Time_ as our `Variables` option. 

1. The third quartile shows the value of 12.305 (days). Thus, approximately 75-78% of the reclaims are handled within 15 days.


### 3. Is the distribution symmetrical?
**Solution:** In Minitab, go to `Graph` >> `Histogram...`. In the `Histogram` popup diaglog, select `With Fit` as our option. Then, in the `Histogram: With Fit` dialog, select _Total Time_ as our `Graph Variables`.

According to the histogram graph, _Total Time_ has a skewed distribution.

Tip: Additionally, we can use perform a normality test to determine whether _Total Time_ is normally distribute. To do so, go to `Stat` >> `Basic Statistics` >> `Normality Test...`. In the `Normality Test`popup dialog, select _Total Time_ as our `Variable`.

When we looked up at the **AD** parameter, which stands for the Anderson-Darling test result, its value was significantly high. Therefore, _Total Time_ does not have a normal distribution.

### 4. Which type of reclaim occurs most often?


**Solution:** In Minitab, go to `Graph` >> `Pie Chart...`. In the `Pie Chart` popup diaglog, select `Chart counts of unique values` as our option, then _Type_ as our `Categorical Variables`.

According to the pie chart, the _OX_ type occurs the most often, which constitutes slightly more than 30%.
