## Questions: What percentage of the patients stays at most 10 days?
**Solution:** Our workflow is as follows: _Length of stay_ &#8594; Numerical (as in Data Type)
&#8594;Type of Distribution then Cumulative Probability.

In Minitab, go to `Graph` >> `Probability Plots...`. In the `Probability Plots` popup dialog, select `Single` as our option. In the dialog, select our `Length of stay` as our `Graph variables`. Then, in the `Distribution` section, we will test what type of distribution our data are by changing our selected distribution one after the other (Normal &#8594; Lognormal &#8594; Weibul).

Among all of these probability plots, we saw the lognormal fited the best due to having the lowest **AD** value. Hence, we can conclude that the _Length of stay_ has the lognormal distribution.

In Minitab, go to `Graph` >> `Empirical CDF...`. In the `Empirical CDFs` popup dialog, select `Single` as our option. In the dialog, select _Length of stay_ as our `Graph variables` option. Then, in the `Distribution` section, we selected lognormal as our distribution for _Length of stay_.

When the graph is created, we opened the graph editor by double-clicking the graph. In the `Edit Graph` dialog, go to `Add` >> `Percentile Lines`. In the dialog, we added 10 (since we considered from 10 days downwards) for our `Show percentile lines at data values`.

According to the graph, we see that 47.2% of the patients stay at most 10 days.
