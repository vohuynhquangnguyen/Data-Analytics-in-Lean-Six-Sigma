## Questions: Are all shifts equally productive?
**Solutions**
Our workflow is as follows: 

Identify what are our responses (CTQ, which are our Ys) and influencing factors (our Xs)  Select the approriate test  Organize our data if necessary  Conduct the test.


Because:
1. Our task is to determine whether our shifts are equally productive.
2. Our CTQ is produced goods (Numerical), and our influencing factor is workshift (Categorical).

The most approriate test is ANOVA (analysis of variance). Hence, our hypothesis is as follows:
1. **Null hypothesis ($H_o$)**: there is no  relation between Xs and Ys
2. **Alternative hypothesis ($H_a$)**: there is a relation between Xs and Ys.

Because our raw data was not in our desired structure, we have to organize them. To do so, go to `Data` >> `Stack` >> `Columns`. In the popup dialog, select _Shift 1_, _Shift 2_, and _Shift 3_ for `Stack the following columns`. Then, store our stacked data in the same worksheet by adding `Column of current worksheet:` and `Store subscripts in:`.


To conduct our ANOVA, go to `Stat` >> `ANOVA` >> `One way...`. In the popup dialog, first select `Response data are in one column for all factor levels` as our option, next select _Productivity_ as our `Response` and _Shift_ as our `Factor`. Then, go to the `Graphs...` section, select `Interval plot`and `Individual value plot`. Finally, go to the `Options...` section, dis-select `Assume equal variances`.

According to the ANOVA's result, the **p-value** is 0.805, which is significantly greater than 0.05. Besides, both the **R-sq** and **R-sq(adj)** are significantly small (<1%). 

On the other hand, when considering the residual plots, we see that there are two outliers of residual, plus the residuals are not normally distributed. This means the ANOVA are not reliable because all the assumptions to make ANOVA reliable are now invalid. Hence, we must conduct the Kruskal-Wallis analysis. To do so, go to `Stat` >> `Nonparametrics` >> `Kruskal-wallis...`. In the popup dialog, select _Productivity_ as our `Response` and _Shift_ as our `Factor`.  

According to the Kruskall-Wallis analysis' result, the **p-value** is equal 0.05. Hence, we can conclude that not all shifts are equally productive because there is an evidence to reject our null hypothesis: _Shift 3_ is by far the most productive.
